# AXES,LABEL,LEGEND,GRID 

```python

days=[1,2,3,4,5,6,7]#Days of week
max_t=[50,51,52,48,47,49,46]#Maximum temperature
min_t=[43,42,40,44,33,35,37]#minimum temperature
avg_t=[45,48,48,46,40,42,41]#average temperature

#PLOTTING
plt.plot(days,max_t)
plt.plot(days,min_t)
plt.plot(days,avg_t)

plt.xlabel('DAYS')
plt.ylabel('TEMPERATURE')

plt.title('WEATHER')
```

```python
days=[1,2,3,4,5,6,7]#Days of week
max_t=[50,51,52,48,47,49,46]#Maximum temperature
min_t=[43,42,40,44,33,35,37]#minimum temperature
avg_t=[45,48,48,46,40,42,41]#average temperature

plt.xlabel('DAYS')
plt.ylabel('TEMPERATURE')

plt.title('WEATHER')

plt.plot(days,max_t,label='max')
plt.plot(days,min_t,label="min")
plt.plot(days,avg_t,label="avg")
#plt.legend(loc='upper right')
plt.legend(loc='best',shadow=True,fontsize='small')
#plt.legend(loc='best',shadow=True,fontsize='large')
```

# BAR CHART

```python
import numpy as np

#BAR FUNCTION DOES NOT SUPPORT THE STRINGS.Hence perform ypos as done

company=['GOOGL','AMZN','MSFT','FB']
revenue=[90,136,89,27]
ypos=np.arange(len(company))
plt.bar(ypos,revenue)
```

Using xticks function

```python
plt.xticks(ypos,company)
plt.bar(ypos,revenue,label='revenue')
plt.xlabel('REVENUE')
plt.ylabel('COMPANY')
plt.title('BAR CHART')
plt.legend(shadow=True)
```

Multiple Bars showing revenue and profit of major US tech companies


```python
company=['GOOGL','AMZN','MSFT','FB']
revenue=[90,136,89,27]
profit=[40,2,34,12]

ypos=np.arange(len(company))
plt.bar(ypos,revenue)

plt.xticks(ypos,company)
plt.bar(ypos,revenue,label='revenue')
plt.bar(ypos,profit,label='profit',width=0.5)
plt.xlabel('REVENUE')
plt.ylabel('COMPANY')
plt.title('BAR CHART')
plt.legend(shadow=True)
```

HORIZONTAL BAR CHART

```python
plt.barh(ypos,revenue,label='revenue')
```

# HISTOGRAMS

```python
blood_sugar = [113, 85, 90, 150, 149, 88, 93, 115, 135, 80, 77, 82, 129]
plt.hist(blood_sugar)

# For the below histogram there are 10 bins(range ) by default(specified in output)
```

USING BINS COLOR AND RWIDTH PARAMETER

```python
blood_sugar = [113, 85, 90, 150, 149, 88, 93, 115, 135, 80, 77, 82, 129]
plt.hist(blood_sugar,bins=3,rwidth=0.5,color='red')
```

PASSING MULTIPLE DATA FOR THE HISTOGRAM

```python
blood_sugar_men = [113, 85, 90, 150, 149, 88, 93, 115, 135, 80, 77, 82, 129]
blood_sugar_women = [67, 98, 89, 120, 133, 150, 84, 69, 89, 79, 120, 112, 100]

plt.hist([blood_sugar_men,blood_sugar_women],rwidth=0.5,bins=3,label=('men','women'),color=('green','pink'))
plt.legend(shadow=True,loc="best")
plt.xlabel('SUGAR LEVEL')
plt.ylabel('TOTAL NO OF PATIENTS')
plt.title('SUGAR TEST')

```

PLOTTING HISTOGRAMS HORIZONTALLY

```python
blood_sugar = [113, 85, 90, 150, 149, 88, 93, 115, 135, 80, 77, 82, 129]
plt.hist(blood_sugar,bins=3,rwidth=0.5,color='red',orientation='horizontal')
```

# Draw pie chart to track down home expenses


```python
exp_vals = [1400,600,300,410,250]
exp_labels = ["Home Rent","Food","Phone/Internet Bill","Car ","Other Utilities"]
plt.pie(exp_vals,labels=exp_labels)
```

RADIUS PARAMETER  AND AXIS FUNCTION

```python
plt.axis("equal")
plt.pie(exp_vals,labels=exp_labels,radius=1.5)
```

AUTOPCT FUNCTION FOR PERCENTAGE

```python
plt.axis("equal")
plt.pie(exp_vals,labels=exp_labels,radius=1.5,autopct='%0.1f%%',shadow=True)
```

EXPLODE FEATURE IN PIECHART

```python
plt.pie(exp_vals,labels=exp_labels,radius=1,autopct='%0.1f%%',shadow=True,explode=[0,0,.5,0,0])
```
